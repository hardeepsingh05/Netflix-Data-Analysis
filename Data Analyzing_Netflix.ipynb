### 6. Handle Missing Values

```python
# Detecting the Missing Values
```

```python
df.isnull().sum()
```

```python
# Let us first see the missing from lower in number to higher in number
```

#### Missing values in 'duration' column

```python
# First going fot 'duration' columns
df[df.isnull()['duration']==True]
```

```python
# Here we see that there is a mismatch occuring in 'rating' and 'duration' columns.
# So, let's swap them
```

```python
index = list(df[df.isnull()['duration']==True].index)
df.loc[index,'duration'] = df.loc[index,'rating']
df.loc[index,'rating'] = np.nan
```

```python
# So, the missing values of duration column has been recovered
# Now, Let's move to the next column who has missing values
```

```python
df.isnull().sum()
```

#### Missing values in 'Rating' column

```python
# Let us first see which values in 'rating' column are missing
df[df.isnull()['rating']==True]
```

```python
df['rating'].value_counts()
```

```python
# We will using 'bfill' method to fill the missing values
# The “bfill” method fills the missing values backward so they are replaced with the next value.
# We can also limit the number of missing values replaced with this method. If we set the limit parameter as 1, then a missing value can only be replaced with its next value. The second or third following value will not be used for replacement.
```

```python
df['rating'] = df['rating'].fillna(method='bfill',limit=1)
```

```python
df['rating'].isnull().sum()
```

```python
# So, we are dealt with the missing values of 'rating' column
# Now, let's go to the next
```

```python
df.isnull().sum()
```

#### Handling missing values in 'date_added' column

```python
df[df.isnull()['date_added']==True]
```

### 7. Data Pre-processing

```python
df['type'].value_counts()
```

```python
#pd.set_option('display.max_rows', None)
df['release_year'].value_counts()[1:21]
```

```python
df.head()
```

```python
df.tail()
```

```python
df1= df.copy(deep=True)
df1.head()
```

```python
# Extracting Months from the 'Date_added' column
l1 = []  
for i in df1['date_added']:
    i = str(i)
    l1.append(i.strip().split(' ')[0])
df1.insert(loc = 7, column = 'month', value = l1)
```

```python
df1.head()
```

```python
from datetime import datetime
df1['date_added']= df1['date_added'].apply(lambda x : x.strip())
x = df1['date_added'].apply(lambda x : datetime.strptime(x, '%B %d, %Y'))
```

```python
# Extracting day of the month
l =[]
for i in x:
    i=str(i)
    l.append(i.strip().split("-")[2])
l1= []
for i in l:
    i = str(i)
    l1.append(int(i.split(" ")[0]))   
```

```python
df1.insert(loc=8,column='day of the month',value=l1)
```

```python
df1.head()
```

```python
# Changing the Duration of the Movie/TV Show, according to the need so that we can visualize it.
list1 = []
for i in df1['duration']:
    i = str(i)
    if 'min' in i:
        x1 = int(int(i.split(' ')[0])/10)
        list1.append(str(x1) + '-' + str(x1 + 1))
    else:
        list1.append(int(int(i.split(' ')[0])))

```

```python
df1.insert(loc = 12, column = 'Movie/TVshow dura', value = list1)
df1.head()
```

```python
df1['type'].value_counts()
```

```python
df1.rename(columns={'type':'Types'},inplace=True)
```

### 8. Univariate Analysis

```python
plt.figure(figsize=(10,4))
x = sns.countplot(data=df1,x='Types',palette='bright')
plt.xlabel('Type of Entertainment')
plt.ylabel("Count of different types of Entertainment")
plt.title("Overview of different type of Entertainments")
for container in x.containers:
    x.bar_label(container)
plt.show()
```

```python
d= df1['Types'].value_counts()
labels = ['Movie','TV Show']
plt.pie(d,labels=labels, autopct='%.0f%%')
plt.title("Type of Enterntaiments (%)")
plt.show()
```

```python
# Splitting the Duration of the Movies and TV Shows
list1 = list(df1['Movie/TVshow dura'])
TV_shows=[]
Movies=[]
for i in list1:
    i = str(i)
    if '-' in i:
        Movies.append(i)
    else:
        TV_shows.append(i)  
```

```python
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(12,5))
x = sns.countplot(Movies,palette = 'bright')
sns.set_theme(font_scale=0.8)
plt.xlabel('Duration of Movies (converted to 9-10 from 90-100 minutes)')
plt.ylabel("Count of different Movie Durations")
plt.title("Duration of Movies")
for container in x.containers:
    x.bar_label(container)


plt.figure(figsize=(12,5))
y = sns.countplot(TV_shows,palette = 'bright')
plt.xlabel('Duration of TV-Shows (in no. of seasons)')
plt.ylabel("Count of seasons of TV-Shows")
plt.title("Duration of TV-Shows")
for container in y.containers:
    y.bar_label(container)
plt.show()
```

```python
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(10,5))
x = sns.countplot(data = df1[df1['Types']=='Movie'],x ='rating',palette = 'bright')
plt.xlabel('Types of Ratings')
plt.ylabel("Count of Ratings")
plt.title("Overview of Different Ratings in Movies")
for container in x.containers:
    x.bar_label(container)


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(10,5))
y = sns.countplot(data = df1[df1['Types']=='TV Show'],x ='rating',palette = 'bright')
plt.xlabel('Types of Ratings')
plt.ylabel("Count of Ratings")
plt.title("Overview of Different Ratings in TV-Shows")
for container in y.containers:
    y.bar_label(container)
plt.show()
```

```python
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(10,5))
z = sns.countplot(data = df1,x ='month',palette = 'bright')
plt.xlabel('Months')
plt.ylabel("Count of Movies & TV-Shows in Months")
plt.title("Overview of Movies & TV-Shows Month Wise")
for container in z.containers:
    z.bar_label(container)


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(10,5))
x = sns.countplot(data = df1[df1['Types']=='Movie'],x ='month',palette = 'bright')
plt.xlabel('Months')
plt.ylabel("Count of Movies in Months")
plt.title("Overview of Movies Month Wise")
for container in x.containers:
    x.bar_label(container)


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(10,5))
y = sns.countplot(data = df1[df1['Types']=='TV Show'],x ='month',palette = 'bright')
plt.xlabel('Months')
plt.ylabel("Count of TV-Shows in Months")
plt.title("Overview of TV-Shows Month Wise")
for container in y.containers:
    y.bar_label(container)
plt.show()
```

```python
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(12,5))
z = sns.countplot(data = df1,x ='day of the month',palette = 'bright')
plt.xlabel('Days')
plt.ylabel("Count of Movies & TV-Shows in Days")
plt.title("Overview of Movies & TV-Shows Day Wise")
for container in z.containers:
    z.bar_label(container)


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(12,5))
x = sns.countplot(data = df1[df1['Types']=='Movie'],x ='day of the month',palette = 'bright')
plt.xlabel('Days')
plt.ylabel("Count of Movies in Days")
plt.title("Overview of Movies Day Wise")
for container in x.containers:
    x.bar_label(container)


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(12,5))
y = sns.countplot(data = df1[df1['Types']=='TV Show'],x ='day of the month',palette = 'bright')
plt.xlabel('Months')
plt.ylabel("Count of TV-Shows in Days")
plt.title("Overview of TV-Shows Day Wise")
for container in y.containers:
    y.bar_label(container)
plt.show()
```

### 9. Bivariate Analysis

```python
# First of all, Let's get the top 30 years based on Count of Movies/TV-Shows
Release_year = df1['release_year'].value_counts()[0:30].rename_axis('Year').reset_index(name='Counts')
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(13,6))
x = sns.barplot(data = Release_year,x='Year',y='Counts',palette = 'bright')
plt.xlabel('Years')
plt.ylabel("Count of Movies/TV-Show")
plt.title("Overview of Movies/TV-Shows released by year")
for container in x.containers:
    x.bar_label(container)
plt.show()
```

```python
sns.set_theme(font_scale=0.8)
plt.figure(figsize=(13,6))
sns.boxplot(data = df1,x='rating',y='day of the month',palette = 'bright')
plt.xlabel('Ratings')
plt.ylabel("Days in a Month")
plt.title("Overview of Movies & TV-Shows by ratings using Boxplot")


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(13,6))
sns.boxplot(data = df1[df1['Types']=='Movie'],x='rating',y='day of the month',palette = 'bright')
plt.xlabel('Ratings')
plt.ylabel("Days in a Month")
plt.title("Overview of Movies by ratings using Boxplot")


sns.set_theme(font_scale=0.8)
plt.figure(figsize=(13,6))
sns.boxplot(data = df1[df1['Types']=='TV Show'],x='rating',y='day of the month',palette = 'bright')
plt.xlabel('Ratings')
plt.ylabel("Days in a Month")
plt.title("Overview of TV-Shows by ratings using Boxplot")
plt.show()
```
